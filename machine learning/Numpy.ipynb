#List vs NumPy array time taken

```python
from time import process_time
```

```python
my_list = [i for i in range(10000)]

start_time = process_time()

my_list = [i+5 for i in my_list]

end_time = process_time()

print(end_time  - start_time)
```

```python
np_array = np.array([i for i in range(10000)])

start_time = process_time()

np_array +=5

end_time = process_time()

print(end_time - start_time)
```

#Opeation on an array

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]

print(list1+list2) # concatination of list
```

```python
# for numpy array

r_arr1 = np.random.randint(0,10,(3,3))
r_arr2 = np.random.randint(10,20,(3,3))

print(r_arr1)
print(r_arr2)
```

```python
print(r_arr1+r_arr2)
print(r_arr1-r_arr2)
print(r_arr1*r_arr2)
print(r_arr1/r_arr2)


```

```python
r_arr1 = np.random.randint(0,10,(3,3))
r_arr2 = np.random.randint(10,20,(3,3))

print(r_arr1)
print(r_arr2)
```

```python
#another way of doing same thing
print(np.add(r_arr1,r_arr2))
print(np.subtract(r_arr1,r_arr2))
print(np.multiply(r_arr1,r_arr2))
print(np.divide(r_arr1,r_arr2))
```

#Array Manipluation

```python
arr_manu = np.random.randint(10,20,(3,3));
print(arr_manu)
print(arr_manu.shape)
```

```python
#transpose of array
arr_manu2= np.transpose(arr_manu)
print(arr_manu2)
```

```python
trans = arr_manu.T
print(trans)
```

```python
#reshaping of array
reshape_arr = np.random.randint(10,20,(2,3));
print(reshape_arr)
print(reshape_arr.shape)
```
