```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set() # this will set the default parameter for visualization
```

**quiver funtion is used to plot the vector**

1st parameter define from where the vector start from x-axis

2nd parameter define from where the vector start from y-axis

3nd define the destination x-axis coordinate

4th define the destination y-axis coordinate

```python
plt.quiver(0,0,4,5)
plt.show()
```

0 and 0: These are the x and y coordinates of the starting point of the arrow.

4 and 5: These are the x and y components of the vector, which determine the length and direction of the arrow.

scale_units='xy': This parameter specifies how the scaling of the arrow lengths should be interpreted. In this case, 'xy' indicates that the arrow lengths are in the same data units as the x and y coordinates.

angles='xy': This parameter specifies how the angles of the arrows should be interpreted. 'xy' indicates that the angles of the arrows are determined by the x and y components of the vector.

scale=1: This parameter determines the scaling factor for the arrow lengths. A value of 1 means that the arrows will have lengths proportional to the vector magnitudes.

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-3,-6,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
