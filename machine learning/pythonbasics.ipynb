## **set**

1. set are formed using curly brackets{}
2. set can have multiple type of value
3. we cannot print the values using indexing
4. we can convert list into set
5. set doesnot allow duplicate values

```python
# set are formed using curly brackets{}

my_set = {1,2,3,5,3}
print(my_set)
```

```python
# set can have multiple type of value

my_set = {1,2.3,5,"english"}
print(my_set)
```

```python
# we cannot print the values using indexing
#print(my_set[0])
```

```python
#set doesnot allow duplicate values

my_set = {1,2,3,1,2,3,1,2,3}
print(my_set)
```

## **DICTIONARY**

```python
#factorial of number
number = int(input("entet the number :"))

factorial =1;

if(number ==0):
  print("factorial of 0 is 1")

for i in range(1 , number+1):
  factorial = factorial *i;

print("factorial of ",number," is : ",factorial)
```

Function in python

```python
def fact(num):
  factorial = 1
  if(num==0):
    return factorial
  else:
    for i in range(1,num+1):
      factorial = factorial*i

  return factorial


print(fact(int(input("enter the number"))))
```
