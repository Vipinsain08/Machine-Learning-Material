#seaborn


* Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures. Seaborn helps you explore and understand your data.




```python
# importing the libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has some built in datasets

```python
tips = sns.load_dataset('tips') # loading the dataset total bill vs tip
```

```python
tips.head()
```

```python
#setting a theme in seaborn
sns.set_theme()
```

```python
#visulaize the data
sns.relplot(data =tips,x='total_bill',y='tip',col = 'time',hue='smoker',style='smoker',size='size')
```

```python
#loading the another data set
irs = sns.load_dataset('iris')
```

```python
irs.head()
```

```python
sns.scatterplot( data= irs, x='sepal_length', y='petal_length', hue='species')
```

```python
sns.scatterplot(data=irs,x='sepal_width',y='petal_width',hue='species')
```
