Handling the  imbalanced data:
An imbalanced data is data which have unequal class distribution

```python
#importing the libraries
import pandas as pd
import numpy as np
```

```python
#lading the dataset into the pandas dataframe
credit_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_data.head()
```

```python
credit_data.tail()
```

```python
#checking the distribution of two classes
credit_data['Class'].value_counts()
```

```python
credit_data.shape
```

This is highly imbalanced data

0 --> legal transaction

 1--> fraud transaction

```python
#separating the legal and fraud transaction

legal_t = credit_data[credit_data.Class == 0]
fraud_t = credit_data[credit_data.Class == 1]
```

```python
print(legal_t.shape)
print(fraud_t.shape)
```

Under-sampling case

build a sample data having similar distribution of transaction

No of fraud transaction:25

```python
sample_data = legal_t.sample(n=25)
```

```python
print(sample_data.shape)
```

Concat the two data frames

```python
new_dataset = pd.concat([sample_data,fraud_t],axis=0)
```
