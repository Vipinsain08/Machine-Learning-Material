***central tendencies***
1. Mean : average value
2. Median : middle values
3. Mode : most frequent value

```python
#analyzing the distribution of salary
sns.distplot(df.salary)
```

this curve is called skew curve in which most of dataset is distribute in one side , so for this case we can not use the mean to replace the missing values as we have outliers which increase the mean value unnecessarly

```python
# replacing the missing values with median value
df['salary'].fillna(df['salary'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
# replacing the missing values with mean value
#df['salary'].fillna(df['salary'].mean(),inplace=True)
```

```python
# replacing the missing values with mode value
df['salary'].fillna(df['salary'].mode(),inplace=True)
```
